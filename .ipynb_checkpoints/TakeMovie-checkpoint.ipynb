```python
from tkinter.constants import ACTIVE, NO
#from env import ObjectTransport
import env
import model 
import Config

import importlib
importlib.reload(model)
importlib.reload(env)
importlib.reload(Config)

import torch
import numpy as np

import matplotlib.colors as colors
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
import moviepy.editor as mp
import matplotlib.patches as patches
import matplotlib.animation as animation


device = torch.device("cpu")
def obs_to_torch(obs: np.ndarray) -> torch.Tensor:
    # scale to `[0, 1]`
    return torch.tensor(obs, dtype=torch.float32, device=device)

def option():
    plt.rcParams['font.family'] = 'Times New Roman' # font familyの設定
    plt.rcParams['mathtext.fontset'] = 'stix' # math fontの設定
    plt.rcParams["font.size"] = 20 # 全体のフォントサイズが変更されます。
    plt.rcParams['xtick.labelsize'] = 15 # 軸だけ変更されます。
    plt.rcParams['ytick.labelsize'] = 15 # 軸だけ変更されます

    
class Test():
    def __init__(self, _env, _model):        
        self.env = _env(0, 0.25, 1)
        self.model = _model
        self.imgs = []
        self.traj = np.zeros([Config.SIMSTEP+1, Config.ROBOT_NUM + 1, 2])
        self.env.reset()
        
        
        option()
        self.fig = plt.figure(figsize=(8,8))
        self.ax = plt.axes()
        

    def run(self):
        obs, *_ = self.env.reset()
        logs = np.zeros([129, 3]) # dis, dis_std, orn
        i = 0
        while True:
            pi, *_ = self.model(obs_to_torch(obs))
            #actions = pi.sample().detach().numpy()

            probs = pi.probs
            probs = probs.detach().numpy()
            actions = np.argmax(probs,axis=1)
            
            self.draw(i)
            
            self.traj[i, 0:Config.ROBOT_NUM, :] = np.array([
                robot.pos[0:2] for robot in self.env.robotsList
            ])
            self.traj[i, -1, :] = np.array([
                obj.pos[0:2] for obj in self.env.objectsList
            ])
            obs, r, done ,info = self.env.step(actions)
            
            i = i + 1
            
            if done:
                self.draw(i)
                self.traj[i, 0:Config.ROBOT_NUM, :] = np.array([
                    robot.pos[0:2] for robot in self.env.robotsList
                ])
                self.traj[i, -1, :] = np.array([
                    obj.pos[0:2] for obj in self.env.objectsList
                ])
                #filename = f'figs/{Config.ROBOT_NUM}agent_6toubun_m0.5.gif'
                #self.imgs[0].save(filename,
                #               save_all=True, append_images=self.imgs[1:], 
                #               optimize=False, duration=100, loop=0)
                
                #clip = mp.VideoFileClip(filename)
                #clip.write_videofile(f"{filename}.mp4")
                #clip.close()
                
                break
                
    def draw(self, i):
        img = self.env.getCameraImage()
        img = Image.fromarray(img)
        #img.save("figs/init_position.png")
        
        imagesize = img.size        # img.size[0]は幅、img.size[1]は高さを表す
        draw = ImageDraw.Draw(img)  # ImageDrawオブジェクトを作成

        #font = ImageFont.truetype("arial.ttf", 64)  # フォントを指定、64はサイズでピクセル単位
        font = ImageFont.truetype('/System/Library/Fonts/Times.ttc', 40)

        # 画像右下に'Sampleと表示' #FFFは文字色（白）
        draw.text((0, 0), f"STEP {i}", font=font, fill="#000")
        self.imgs.append(
            img
        )
        #plt.cla()
        #plt.imshow(img)

        #plt.draw()
        #plt.pause(0.01)
        
    
    def draw(self,step):
        ax = self.ax
        fig = self.fig
        ax.cla()
        
        env = self.env
        
        
        
        robot_colors = list(colors.TABLEAU_COLORS.values())
        
        
        for i in range(Config.ROBOT_NUM):
            #print(i)
            ax.plot(self.traj[0:step, i, 0], self.traj[0:step, i, 1], color=robot_colors[0])
        
        #ax.plot(self.traj[0:step, -1, 0], self.traj[0:step, -1, 1], color='#98623C')
        
        for i in range(Config.ROBOT_NUM):
            ax.scatter(self.traj[0, i, 0], self.traj[0, i, 1], c=robot_colors[0])
        
        
        obs_pos = env.objectsList[0].pos
        c = patches.Circle(xy=obs_pos, radius=0.4, fc='#98623C', ec='k')
        ax.add_patch(c)
        
        for i, robot in enumerate(env.robotsList):
            pos = robot.pos
            phi = robot.phi
            
            ax.plot([ pos[0]+0.02*np.cos(phi), pos[0]+0.05*np.cos(phi)], [pos[1]+0.02*np.sin(phi), pos[1]+0.05*np.sin(phi)], linewidth=2, color="k")
            c = patches.Circle(xy=pos, radius=0.05, fc=robot_colors[0], ec='k')
            ax.add_patch(c)

            #ax.text(pos[0]+0.05,pos[1]+0.05, f"{i+1}", size=25)
        
        #ax.text(obs_pos[0]-1.15,obs_pos[1]+1.1, f"Step {step}", size=20)
        #ax.set_xlim([obs_pos[0]-1.2,obs_pos[0]+1.2])
        #ax.set_ylim([obs_pos[1]-1.2,obs_pos[1]+1.2])

        #ax.set_xticks(obs_pos[0] + np.arange(-1,1.5,0.5))
        #ax.set_yticks(obs_pos[1] + np.arange(-1,1.5,0.5))
        
        ax.tick_params(direction='in', length=0, width=2, color='gray')
        
        ax.set_aspect('equal')
        #ax.set_xlabel("x [m]", fontsize=16)
        #ax.set_ylabel("y [m]", fontsize=16)
        
        ax.tick_params(labelbottom=False,
               labelleft=False,
               labelright=False,
               labeltop=False)
        
        
        #plt.savefig(f"movie/{step}.pdf", bbox_inches="tight", pad_inches=0.05)
        
        fig.canvas.draw()
        im = np.array(fig.canvas.renderer.buffer_rgba())
        
        img = Image.fromarray(im)
        #img = Image.frombytes('RGB', self.fig.canvas.get_width_height(), self.fig.canvas.tostring_rgb())
        self.imgs.append(img) 
```

```python


_model = model.Model(Config.PolicySetting, Config.ValueSetting)
_model.load(f"model/model_1000_4_2.pth")

test = Test(env.ObjectTransport, _model)
test.run()
```
